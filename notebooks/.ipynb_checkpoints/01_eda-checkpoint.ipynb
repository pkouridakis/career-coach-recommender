# Exploratory Data Analysis — Career Coach Recommender

Before building anything, I want to understand the data.
The goal here is to answer a few simple questions:
1. What certificates exist and how are they structured?
2. What have users already bought and who are they?
3. Are there common learning paths?
4. What do the interaction signals tell us?

These insights will directly inform how I build the recommender.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all four datasets
catalog   = pd.read_csv('../data/catalog.csv')
users     = pd.read_csv('../data/users.csv')
purchases = pd.read_csv('../data/purchases.csv')
signals   = pd.read_csv('../data/signals.csv')

print('catalog:  ', catalog.shape)
print('users:    ', users.shape)
print('purchases:', purchases.shape)
print('signals:  ', signals.shape)

print('Columns Types: ')
print(catalog.dtypes)
print(users.dtypes)
print(purchases.dtypes)
print(signals.dtypes)
```

## 1. The Certificate Catalog

Let's understand what we're recommending — the 12 available certificates.

```python
# Quick look at the catalog
catalog[['cert_id', 'name', 'family', 'level', 'prerequisites', 'price']]
```

```python
# Which certs have no prerequisites? These are entry points.
entry_points = catalog[catalog['prerequisites'].isna()]
print('Entry-point certificates (no prerequisites):')
print(entry_points[['cert_id', 'name', 'family']].to_string(index=False))
```

```python
# Skills are pipe-separated — let's see all unique skills across the catalog
all_skills = catalog['skills'].str.split('|').explode().str.strip()
print(f'Total unique skills in catalog: {all_skills.nunique()}')
print(all_skills.value_counts().head(10))
```

## 2. User Profiles

Who are the learners? Understanding their roles and goals helps us know what to recommend.

```python
users.head(5)
```

```python
# Role distribution
role_counts = users['role'].value_counts()

role_counts.plot(kind='barh', figsize=(8, 5), color='steelblue')
plt.title('Users by Role')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Career goal distribution
goal_counts = users['goal'].value_counts()

goal_counts.plot(kind='barh', figsize=(8, 4), color='coral')
plt.title('Users by Career Goal')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

What have people actually bought? This reveals real learning paths.

```python
# Which certificates are most purchased?
popular = purchases['cert_id'].value_counts()

popular.plot(kind='bar', figsize=(10, 4), color='mediumseagreen')
plt.title('Most Purchased Certificates')
plt.ylabel('Number of purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# How many certificates does each user buy on average?
certs_per_user = purchases.groupby('user_id')['cert_id'].count()

print(f'Average certs per user: {certs_per_user.mean():.1f}')
print(f'Max certs by one user:  {certs_per_user.max()}')
print(f'Users with only 1 cert: {(certs_per_user == 1).sum()}')

certs_per_user.value_counts().sort_index().plot(kind='bar', figsize=(7,4), color='slateblue')
plt.title('Distribution: How many certs per user?')
plt.xlabel('Number of certificates purchased')
plt.ylabel('Number of users')
plt.tight_layout()
plt.show()
```

```python
# Co-occurrence: which pairs of certs are bought together by the same user?
# This is the foundation of our co-occurrence recommender.

user_certs = purchases.groupby('user_id')['cert_id'].apply(list)

from itertools import combinations
from collections import Counter

pair_counts = Counter()
for cert_list in user_certs:
    for pair in combinations(sorted(cert_list), 2):
        pair_counts[pair] += 1

print('Top 10 most co-purchased certificate pairs:')
for pair, count in pair_counts.most_common(10):
    print(f'  {pair[0]} + {pair[1]} → {count} users')
```

## 4. Interaction Signals

The signals file captures the funnel: impression → click → add_to_cart → purchase.
This tells us which certificates are engaging, not just bought.

```python
signals.head()
```

```python
# CTR (Click-Through Rate) per certificate
# CTR = clicks / impressions — how often a cert gets clicked when shown

impressions = signals[signals['event'] == 'impression'].groupby('cert_id').size()
clicks      = signals[signals['event'] == 'click'].groupby('cert_id').size()

ctr = (clicks / impressions).fillna(0).sort_values(ascending=False)

print("CTR per certificate:")
print(ctr)
print(f"\nBest performing cert: {ctr.idxmax()} ({ctr.max():.2%})")
print(f"Worst performing cert: {ctr.idxmin()} ({ctr.min():.2%})")

ctr.plot(kind='bar', figsize=(10, 4), color='tomato')
plt.title('Click-Through Rate (CTR) per Certificate')
plt.ylabel('CTR')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# How strong is each signal type as a purchase predictor?
# We'll use this intuition to assign weights in the ranking step.

# Idea: impression=1, click=2, add_to_cart=3, purchase=4
# A user who added to cart is more likely to buy than one who just clicked.

event_counts = signals.groupby(['user_id', 'cert_id', 'event']).size().unstack(fill_value=0)
print(event_counts.head(8))
```

```python
import sys
sys.path.append('../src')
from recommender import recommend, users_list, cert_list, purchases_list, signals_list, co_occurrence_recommend
from data_loader import load_catalog, load_users, load_purchases, load_signals

catalog   = load_catalog()
users     = load_users()
purchases = load_purchases()
signals   = load_signals()

recommendations = recommend('U0002',users, catalog, purchases, signals)

purchased_list = purchases[purchases['user_id'] == 'U0002']["cert_id"].tolist()

def recall_calc(purchased_list, recommendations):
    if not purchased_list:
        return 0.0

    counter = 0
    for cert in set(purchased_list):
        if cert in recommendations:
            counter += 1

    return counter / len(set(purchased_list))

print(recall_calc(purchased_list,recommendations))
print(purchased_list)
pd.DataFrame(recommendations)
```

## 5. Key Takeaways for the Recommender

After exploring the data, here's what I'll use in the recommendation logic:

1. **Prerequisites matter** — 6 out of 12 certs require another cert first. The recommender must enforce this rule.

2. **Learning paths are family-based** — users tend to stay within a family (ITIL Foundation → ITIL Practitioner). Co-occurrence confirms this.

3. **Skills overlap is a strong signal** — users whose skills match a cert's taught skills are more likely to buy it (content-based).

4. **Signals give us intent** — add_to_cart and click are strong signals even without a purchase.

5. **Cold-start is possible** — some users have no purchases yet. For them, we rely only on content-based (skills + goal match).
